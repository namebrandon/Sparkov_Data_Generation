```python
leg_trans = adults_50up_male_urban[adults_50up_male_urban.is_fraud == 0]
fraud_trans = adults_50up_male_urban[adults_50up_male_urban.is_fraud == 1]
```

```python
leg_trans.head()
```

```python
plt.figure(figsize=(20,20))
n_r, n_c = 3, 2

plt.subplot(n_r, n_c, 1)
sns.countplot(leg_trans.trans_date_trans_time.dt.hour).set_title('Reg Trans - Hourly Dist')

plt.subplot(n_r, n_c, 2)
sns.countplot(fraud_trans.trans_date_trans_time.dt.hour).set_title('Fraud Trans - Hourly Dist')

plt.subplot(n_r, n_c, 3)
sns.countplot(leg_trans.trans_date_trans_time.dt.dayofweek).set_title('Reg Trans - day_of_week Dist')

plt.subplot(n_r, n_c, 4)
sns.countplot(fraud_trans.trans_date_trans_time.dt.dayofweek).set_title('Fraud Trans - day_of_week Dist')

plt.subplot(n_r, n_c, 5)
sns.distplot(leg_trans.amt).set_title('Reg Trans - amt Dist')

plt.subplot(n_r, n_c, 6)
sns.distplot(fraud_trans.amt).set_title('Fraud Trans - amt Dist')
```
